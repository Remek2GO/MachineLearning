```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.datasets import make_moons
import seaborn as sns

X, y= make_blobs(n_samples=210, centers=3, n_features=2, cluster_std=0.5, shuffle=True,random_state=0)

plt.scatter(X[:,0], X[:,1], c='red', marker='x')
plt.title('Wizualizacja przykładów dla 3 klastrów')
plt.xlabel('X0')
plt.ylabel('X1')
plt.show()
```

Zadanie 1

```python
km = KMeans(n_clusters=3, init='random', random_state=0, n_init="auto")
km.fit_predict(X)

plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], s=70, c=km.labels_ , cmap=plt.cm.prism) 
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.xlabel('X0')
plt.ylabel('X1')

inertia = km.inertia_
print(f'Wartość zniekształceń wynosi: ', inertia)
```

Zadanie 2

```python
X, y= make_blobs(n_samples=210, centers=5, n_features=2, cluster_std=0.8, shuffle=True,random_state=0)

plt.scatter(X[:,0], X[:,1], c='red', marker='x')
plt.title('Wizualizacja przykładów dla 5 klastrów')
plt.xlabel('X0')
plt.ylabel('X1')
plt.show()

km = KMeans(n_clusters=5, init='random', random_state=0, n_init="auto")
km.fit_predict(X)

inertia = km.inertia_
print(f'Wartość zniekształceń dla tych samych klastrów wynosi: ', inertia)

plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], s=70, c=km.labels_ , cmap=plt.cm.prism) 
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.title('Dopasowanie danych do modelu dla tej samej liczby klastrów')
plt.xlabel('X0')
plt.ylabel('X1')

km_r = KMeans(n_clusters=3, init='random', random_state=0, n_init="auto")
km_r.fit_predict(X)

inertia_r = km_r.inertia_
print(f'Wartość zniekształceń dla różnych klastrów wynosi: ', inertia_r)

plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], s=70, c=km_r.labels_ , cmap=plt.cm.prism) 
plt.scatter(km_r.cluster_centers_[:, 0], km_r.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.title('Dopasowanie danych do modelu dla różnej liczby klastrów')
plt.xlabel('X0')
plt.ylabel('X1')
```

Zadanie 3

```python
X, y= make_blobs(n_samples=210, centers=5, n_features=2, cluster_std=0.8, shuffle=True, random_state=0)

plt.scatter(X[:,0], X[:,1], c='red', marker='x')
plt.title('Wizualizacja przykładów dla 5 klastrów')
plt.xlabel('X0')
plt.ylabel('X1')
plt.show()

km = KMeans(n_clusters=5, init='k-means++', random_state=0, n_init="auto")
km.fit_predict(X)

inertia = km.inertia_
print(f'Wartość zniekształceń dla tych samych klastrów wynosi: ', inertia)

plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], s=70, c=km.labels_ , cmap=plt.cm.prism) 
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.title('Dopasowanie danych do modelu dla tej samej liczby klastrów')
plt.xlabel('X0')
plt.ylabel('X1')

km_r = KMeans(n_clusters=3, init='k-means++', random_state=0, n_init="auto")
km_r.fit_predict(X)

inertia_r = km_r.inertia_
print(f'Wartość zniekształceń dla różnych klastrów wynosi: ', inertia_r)

plt.figure(figsize=(6,5))
plt.scatter(X[:,0], X[:,1], s=70, c=km_r.labels_ , cmap=plt.cm.prism) 
plt.scatter(km_r.cluster_centers_[:, 0], km_r.cluster_centers_[:, 1], marker='*', s=200,
            color='blue', label='Centers')
plt.legend(loc='best')
plt.title('Dopasowanie danych do modelu dla różnej liczby klastrów')
plt.xlabel('X0')
plt.ylabel('X1')
```

Zadanie 4

```python
no_of_clusters = [1, 2, 3, 4, 5, 6, 7, 8]
inertia_list = []
for k in no_of_clusters:  
    km = KMeans(n_clusters=k, init='random', random_state=0, n_init="auto")
    km.fit_predict(X)

    inertia_list.append(km.inertia_)

plt.plot(no_of_clusters, inertia_list, 'bx-')
plt.xlabel("No of clusters")
plt.ylabel("Inertia")
plt.title("Elbow method")
plt.show()
```

```python
x, y = make_moons(1000, noise=.05, random_state=0)
X_moon = pd.DataFrame(x, columns = ['f1','f2'])
#metoda klasteryzacji
# km -model KMeans
#y_km - wynik predykcji

km = KMeans(n_clusters=2, init='random', random_state=0, n_init="auto")
y_km = km.fit_predict(x)
 
#wykres
X_moon['k_means'] = y_km
sns.lmplot(data=X_moon, x='f1', y='f2', fit_reg=False, hue = 'k_means', palette = ['#eb6c6a', '#6aeb6c']).set(title='Algorytm k-srednich')
plt.scatter(km.cluster_centers_[:, 0], km.cluster_centers_[:, 1], c='black', s=100, alpha=0.5)
plt.show()
```
